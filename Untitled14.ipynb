# Ans7-

Analyze the impact of time-related features (e.g., date/time of booking, departure/arrival times).

Use line plots or bar charts to show average prices over time.

```python
# Set the figure size
plt.figure(figsize=(16, 12))

# Bar plot for 'date' vs 'Price'
plt.subplot(2, 2, 1)
flight_data.groupby('date')['price'].mean().plot(kind='bar', color='skyblue')
plt.title('Average Price by Date')
plt.xlabel('Date')
plt.ylabel('Average Price')

# Bar plot for 'dep_time' vs 'Price'
plt.subplot(2, 2, 2)
flight_data.groupby('dep_time')['price'].mean().plot(kind='bar', color='skyblue')
plt.title('Average Price by Departure Time')
plt.xlabel('Departure Time')
plt.ylabel('Average Price')

# Bar plot for 'time_taken' vs 'Price'
plt.subplot(2, 2, 3)
flight_data.groupby('time_taken')['price'].mean().plot(kind='bar', color='skyblue')
plt.title('Average Price by Flight Duration')
plt.xlabel('Flight Duration')
plt.ylabel('Average Price')

# Bar plot for 'arr_time' vs 'Price'
plt.subplot(2, 2, 4)
flight_data.groupby('arr_time')['price'].mean().plot(kind='bar', color='skyblue')
plt.title('Average Price by Arrival Time')
plt.xlabel('Arrival Time')
plt.ylabel('Average Price')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
